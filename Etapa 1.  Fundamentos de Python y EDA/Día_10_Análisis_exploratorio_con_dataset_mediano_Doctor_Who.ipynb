#### ¿Qué porcentaje del total de ratings representa la puntuacion mayor de 7?

#### Para averiguarlo procedemos a contar los ratings que son mayores que 7 y lo dividimos por el número total de ratings y a lo que obtenemos lo multiplicamos por 100

```python
(df[df["Rating"] > 7 ]["Rating"].count() /df["Rating"].count())*100
```

### 📘 Explicación paso a paso:
1. df["Rating"] > 7 → genera una serie booleana que indica qué filas tienen rating mayor que 7.

2. df[df["Rating"] > 7] → filtra el DataFrame para quedarte solo con esas filas.

3. ["Rating"] → selecciona la columna Rating.

4. .count() → cuenta cuántos valores hay (es decir, cuántos ratings son mayores que 7).

5. df["Rating"].count() → cuenta el total de ratings.

6. El cociente da el porcentaje de ratings > 7, que multiplicas por 100.



```python
# Tambien se puede usar

(df["Rating"] > 7).mean()*100

# Esto funciona porque True cuenta como 1 y False como 0, así que .mean() en una serie booleana te da el porcentaje directamente.
```

## Visualizacion básica

#### Pasamos ahora a practicar un poco de visualizaciones y gráficas

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
```

```python
# numero de temporadas
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="Season", order=df["Season"].value_counts().index)
plt.title("Número de temporadas")
plt.xticks(rotation=45)
plt.show()

# lo que vemos en la grafica es que se ha contado cuantas veces sale cada temporada, lo cual nos puede llevar a deducir cuantos capis tiene cada temporada
```

```python
# Promedios agrupados
# Vamos a ver promedios de numero de capitulos por temporada:
comparativas_temporadas = df.groupby("Season")["Episode"].mean().sort_values(ascending=False)
comparativas_duracion = df.groupby("Season")["Duration"].mean().sort_values(ascending=False)
print(comparativas_temporadas)
print(comparativas_duracion)
```

Aqui queria comparar el numero de capitulos medio por temporada, y podemos ver que desde la temporada 1 hasta la 7 habia una media de 7 capitulos justos pero ese numero se ha ido reduciendo conforme avanzaban las temporadas. tambien vemos un aumento de la duracion de los capitulos por temporada a medida que avanzamos. lo que da a entender que aunque el numero de capitulos por temporada a disminuido, no asi e tiempo de visualizacion pues la duracion de los capitulos ha aumentado conforme avanzamos en temporadas

```python
# Transformaciones por grupo (repaso)
# Creamos una nueva columna con el promedio de capitulos por temporada:
df["media_capitulos"] = df.groupby("Season")["Episode"].transform("mean")
df["media_capitulos"]
```

```python
# Creamos tambien una columna con la diferencia respecto al promedio
df["diferencia_capis"] = df["Episode"] - df["media_capitulos"]
df["diferencia_capis"]
```
